```python
import shutil
import pandas as pd
from sqlalchemy import create_engine
import os.path
from common.partition import get_random_partition
```

```python
path = '../../../external/dataset/'
destination = '../../../external/upload/'
```

```python
engine = create_engine(f'sqlite:///{path}metadata.db')
metadata_df = pd.read_sql('ImageData', engine)

# Take only images with captions
images_paths : str = [pair[0][2:] for pair in metadata_df[metadata_df['caption'] != ''][['image_path']].values]
# Adjust path to be linux-like
images_paths = [image_path.replace('\\', '/') for image_path in images_paths]
# Take a random subset from all images
images_paths = get_random_partition(images_paths, 0.4)[0]

# Subset length
len(images_paths)
```

```python
for image_path in images_paths:
    try:
        image_name = os.path.basename(path+image_path)
        os.remove(destination+image_name)
    except Exception as e:
        print(e)
```

```python
for image_path in images_paths:
    try:
        image_name = os.path.basename(path+image_path)
        shutil.copyfile(path+image_path, destination+image_name)
    except Exception as e:
        print(e)
```
