```python
import cv2
import ultralytics
from ultralytics import YOLO
import os
import glob  # Para obtener una lista de archivos

# Carga el modelo YOLO
model = YOLO('runs/detect/train2/weights/best.pt')

# Carpeta que contiene las imágenes
input_folder = 'imagenes_2'  # Reemplaza con la ruta a tu carpeta de imágenes

# Carpeta para guardar las imágenes recortadas
output_folder = 'detected_images'
os.makedirs(output_folder, exist_ok=True)

# Obtiene una lista de todos los archivos .jpg y .jpeg en la carpeta
image_files = glob.glob(os.path.join(input_folder, '*.jpeg')) + glob.glob(os.path.join(input_folder, '*.jpeg'))

# Itera sobre cada archivo de imagen
for image_path in image_files:
    # Carga la imagen
    img = cv2.imread(image_path)

    if img is None:
       print(f"Error: No se pudo leer la imagen: {image_path}")
       continue  # Salta a la siguiente imagen si no se pudo leer

    # Realiza la detección
    results = model.predict(source=img, conf=0.5)

    # Obtiene el nombre de la imagen sin la extensión
    base_name = os.path.splitext(os.path.basename(image_path))[0]

    # Procesa los resultados
    for i, r in enumerate(results):
        boxes = r.boxes
        for j, box in enumerate(boxes):
            x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())
            
            conf = box.conf[0].item()
            cls = box.cls[0].item()
            label = model.names[int(cls)]

            # Recorta la imagen detectada
            cropped_image = img[y1:y2, x1:x2]

            # Guarda la imagen recortada
            filename = f'{base_name}_detected_{i}_{j}_{label}_{conf:.2f}.jpg' # nombre unico de cada imagen recortada
            output_path = os.path.join(output_folder, filename)
            cv2.imwrite(output_path, cropped_image)

    print(f"Procesada la imagen: {image_path}")
print("Proceso completado.")
```
